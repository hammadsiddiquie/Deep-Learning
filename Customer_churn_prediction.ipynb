```python
for col in df:
  if df[col].dtypes == "object":
    print(f"{col}: {df[col].unique()}")
```

```python
df.replace('No internet service', 'No', inplace=True)
df.replace('No phone service', 'No', inplace=True)
```

```python
for col in df:
  if df[col].dtypes == "object":
    print(f"{col}: {df[col].unique()}")
```

```python
for col in df:
  if df[col].dtypes == "object":
    df.replace({'Yes':1, 'No':0}, inplace=True)
```

```python
for col in df:
    print(f"{col}: {df[col].unique()}")
```

```python
df['gender'] = df['gender'].map({'Male': 1, 'Female': 0})
df['gender'].unique()
```

```python
for col in df:
    print(f"{col}: {df[col].unique()}")
```

```python
# # Select categorical columns
# categorical_cols = ['InternetService', 'Contract', 'PaymentMethod']

# Apply one-hot encoding
df = pd.get_dummies(df, columns=['InternetService'])
df = pd.get_dummies(df, columns=['Contract'])
df = pd.get_dummies(df, columns=['PaymentMethod'])
```

```python
for col in df:
    print(f"{col}: {df[col].unique()}")
```

```python
df.dtypes
```

```python
for col in df:
  if df[col].dtypes == "bool":
    df.replace({True:1, False:0}, inplace=True)
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import minmax_scale
df['tenure'] = minmax_scale(df['tenure'])
df['MonthlyCharges'] = minmax_scale(df['MonthlyCharges'])
df['TotalCharges'] = minmax_scale(df['TotalCharges'])
```

```python
for col in df:
    print(f"{col}: {df[col].unique()}")
```

```python
x = df.drop('Churn', axis=1)
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
from tensorflow.keras.layers import LeakyReLU

model = keras.Sequential([
    keras.layers.Dense(64, input_shape=(26,)),
    LeakyReLU(alpha=0.1),
    keras.layers.Dense(32),
    LeakyReLU(alpha=0.1),
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# model = keras.Sequential([
#     keras.layers.Dense(100, input_shape=(26,), activation='relu'),
#     keras.layers.Dense(50, activation='relu'),
#     keras.layers.Dense(25, activation='relu'),
#     keras.layers.Dense(1, activation='sigmoid')
# ])

# model = keras.Sequential([
#     keras.layers.Dense(26, input_shape=(26,), activation='relu'),
#     keras.layers.Dense(1, activation='sigmoid')
# ])

# opt = keras.optimizers.Adam(learning_rate=0.01)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=10)
```

```python
model.evaluate(x_test, y_test)
```

```python
from tensorflow.keras.layers import Dropout

model1 = keras.Sequential([
    keras.layers.Dense(64, input_shape=(26,), activation='relu'),
    Dropout(0.5),
    keras.layers.Dense(32, activation='relu'),
    Dropout(0.5),
    keras.layers.Dense(16, activation='relu'),
    Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

from tensorflow.keras.optimizers import Adam

opt = Adam(learning_rate=0.001)  # Start with 0.001, adjust as needed
model1.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
model1.fit(x_train, y_train, epochs=10)
```

```python
model1.evaluate(x_test, y_test)
```
